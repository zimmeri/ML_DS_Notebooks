## 1. [20 pts] Define each of the following machine learning terms:
### dataset
A data set is a collection of data, and this data can be numerical, categorical, or a combination of both. In most cases, a dataset will have rows of data with different features, and sometimes a classification. Datasets are used to train, test, and validate machine learning models.    
### training, testing, validation dataset
The overall data set can be split into two or three different datasets, depending on the pipeline (using k-fold cross validation may eliminate using an initial validation set). At the minimum, a training and testing set are necessary. A training set is used to develop the initial model. The data is fed into the algorithm so that it can make classifications on later data. The validation set is used to check the accuaracy of the model on different hyperparameters. The testing set is used to grade the final model.
### ground truth, label
The ground truth are the labels in the provided data set. They are human checked and verified as true so that accompanying models can be trained on this data. Labels are the desired classification. In the example, the hand written digit numbers in the dataset get a label based on what number they are. If the picture shows a 9, the ground truth is 9, verified by a human.  Then the model will predict a label based on the testing input.
### pre-processing, feature, numerical, nominal
Pre-processing is arguably the most important part of machine learning. There are a couple of steps to preprocessing. One step in pre-processing is making sure the data is correct. This includes removing invalid data points such as Null, None, or the provided example of February 31st. The next step is formatting the data into a usable format, such as flattening or shaping the data rows. After that, the data needs to remove outliers and be normalized so the model does not overfit to the training data. An optional step is adding more data points through data augmentation. Finally, feature extraction or selection can be used to generalize on only a couple of columns.
### decision surface
A decision surface is the hyperplane a model creates that seperates the classification a model will output. On one side of the surface, a classification will be made, and on the other side, and opposing classification will be made.
### model validation, accuracy, cross-validation
This question seems to be a bit ambiguos because model validation could be how the model is validated (validation set vs cross-validation) or evaluated (accuracy vs f1 score). I'm guessing model validation is more about gauging how well a model performed compared to other models that could be created for the same problem. Accuracy is the number of correct predictions over the total number of predictions, which is not always the best metric for model validation. Cross-validation splits the training set into k samples, then usually trains a model with specific hyperparameters on k-1 samples and validates on 1 sample. This is used to find the best hyperparemeters while not significantly reducing the training data.

### More training helps? Verify by increasing the training dataset size from let's say starting 50% training and then remaining 50% testing, and then 60% training, etc.
In this case, naive bayes does not benefit from more training data because there is too little data to begin with. The problem here is no matter what, the model will be overfit whether 90 or 75 datapoints.
The decision tree benefits from more training data.
### Will the performance plateau? Show it on a plot.
The performance is too unpredicatble to know tell whether performance will plateau.

```python
train50 = irisDataFrame.sample(frac=.5).sort_index()
test50 = irisDataFrame[~irisDataFrame.index.isin(train50.index)]
train60 = irisDataFrame.sample(frac=.6).sort_index()
test40 = irisDataFrame[~irisDataFrame.index.isin(train60.index)]
```

```python
def bayes(train, test):
    clf = MultinomialNB()
    clf.fit(train.iloc[:, 0:4], train.iloc[:, 4])
    predictions = clf.predict(test.iloc[:, 0:4])
    return metrics.accuracy_score(test.iloc[:, 4], predictions)

def decTree(train, test):
    clf = tree.DecisionTreeClassifier()
    clf.fit(train.iloc[:, 0:4], train.iloc[:, 4])
    predictions = clf.predict(test.iloc[:, 0:4])
    return metrics.accuracy_score(test.iloc[:, 4], predictions)
```

```python
#bayes 50-50 split
b50 = bayes(train50, test50)
b60 = bayes(train60, test40)
d50 = decTree(train50,test50)
d60 = decTree(train60,test40)
print(f'Bayes 50-50 split: {b50}\nBayes 60-40 split: {b60}\nDecision Tree 50-50 split: {d50}\nDecision Tree 60-40 split: {d60}')
```

```python
bayesAcc=[]
decAcc=[]
for x in np.arange(.1, 1, .1):
    train =irisDataFrame.sample(frac=x).sort_index()
    test = irisDataFrame[~irisDataFrame.index.isin(train50.index)]
    bayesAcc.append(bayes(train,test))
    decAcc.append(decTree(train,test))
print(f'Bayes Accuracy: {bayesAcc}\nDec Accuracy: {decAcc}')
```

```python
g = sns.lineplot(x= np.arange(.1, 1, .1), y=bayesAcc)
plt.title('Bayes Accuracy')
plt.show()
```

```python
g = sns.lineplot(x= np.arange(.1, 1, .1), y=decAcc)
plt.title('Decision Tree Accuracy')
plt.show()
```
