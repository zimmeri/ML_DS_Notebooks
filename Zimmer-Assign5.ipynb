```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.impute import KNNImputer
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
```

```python
#saving labels for after imputation
train = pd.read_csv('train.csv').drop(columns=['Name', 'PassengerId']) #name is just repetitive nonnumeric passengerID, passengerID is index
test = pd.read_csv('test.csv').drop(columns=['Name', 'PassengerId'])
testLabels = pd.read_csv('gender_submission.csv')
passengerId = testLabels['PassengerId']
testLabels = testLabels.drop(columns=['PassengerId'])
trainLabels = train['Survived']
train = train.drop(columns='Survived')
train
```

```python
def preprocess(data):
    nanCol = []
    stringCol = ~data.columns.isin(data.select_dtypes(exclude=['object']).columns)
    le = LabelEncoder()
    #check to see value counts of columns, also check to see if col has NAN, also check to see if string
    for col in data.columns:
        nanCheck = data[col].isnull()
        if any(nanCheck):
            nanCol.append(True)
        else:
            nanCol.append(False)
        #print(col + ' ' + str(len(data[col].unique()))+'\n', data[col].value_counts(), '\n\n')
    cabinNull = data[data['Cabin'].isnull()].index.tolist() #get null values
    embarkedNull = data[data['Embarked'].isnull()].index.tolist() #get null values
    print(f'Columns with Null: {data.columns[nanCol]}, String Columns: {data.columns[stringCol]}')
    #Encode String columns
    for col in data.columns[stringCol]:
        le.fit(data[col])
        data[col] = le.transform(data[col])
    #convert back to np.nan for imputation
    for index in cabinNull:
        data.loc[index, 'Cabin'] = np.nan
    for index in embarkedNull:
        data.loc[index, 'Embarked'] = np.nan
    # impute missing values
    return data
```

```python
encodedTrain = preprocess(train)
encodedTrain
```

```python
#impute after so that we can keep the same fit for test
knn_imputer = KNNImputer(n_neighbors=5)
knn_imputer.fit(encodedTrain)
imputedTrain = pd.DataFrame(knn_imputer.transform(encodedTrain), columns=encodedTrain.columns)
imputedTrain
```

```python
encodedTest = preprocess(test)
encodedTest
```

```python
imputedTest = pd.DataFrame(knn_imputer.transform(encodedTest), columns=encodedTest.columns)
imputedTest
```

```python
def classifier(train, trainLabels, test, testLabels, depth=100, trees=1000):
    clf = RandomForestClassifier(max_depth=depth, n_estimators = trees, random_state=1)
    clf.fit(train, trainLabels)
    y_pred = clf.predict(test)
    CM = confusion_matrix(testLabels, y_pred)
    score = pd.Series({'Depth': depth, 
                       'Trees': trees, 
                       'Accuracy': accuracy_score(testLabels, y_pred), 
                       'F1': f1_score(testLabels,y_pred, average='weighted'),
                       'TP': CM[0][0], 'TN':CM[1][1], 'FP':CM[0][1], 'FN':CM[1][0] })
    return score, y_pred
```

```python
#data is small which is not great, still going to use k-cross fold to find the best set of hyperparameters
#choosing 4 which might take a while but oh well
kf = KFold(n_splits=4, random_state=13, shuffle=True)
scores = pd.DataFrame(columns=['Depth', 'Trees','Accuracy','F1','TP','TN','FP','FN'])
trees = [1,10,100,500,1000]
depths=[1,10,50,75,100,200,500]
for tree in trees:
    for depth in depths:
        for train_index, test_index in kf.split(train):
            X_train, X_test = imputedTrain.iloc[train_index], imputedTrain.iloc[test_index]
            y_train, y_test = trainLabels.iloc[train_index], trainLabels.iloc[test_index]
            scores = scores.append(classifier(X_train, y_train, X_test, y_test, depth, tree)[0], ignore_index=True)
scores = scores.groupby(['Depth','Trees']).mean().reset_index()
scores
```

```python
def getTarget(data, col, targetCol, aggregator):
    if aggregator == 'min':
        return int(data[col].iloc[data[targetCol].idxmin()])
    if aggregator == 'max':
        return int(data[col].iloc[data[targetCol].idxmax()])
```

```python
#Find the best scoring forests and run them on the test
bestAcc = [getTarget(scores, 'Depth', 'Accuracy', 'max'), getTarget(scores, 'Trees', 'Accuracy', 'max')]
bestF1 = [getTarget(scores, 'Depth', 'F1', 'max'),getTarget(scores, 'Trees', 'F1', 'max')]
bestTP= [getTarget(scores, 'Depth', 'TP', 'max'),getTarget(scores, 'Trees', 'TP', 'max')]
bestTN= [getTarget(scores, 'Depth', 'TN', 'max'),getTarget(scores, 'Trees', 'TN', 'max')]
bestFP = [getTarget(scores, 'Depth', 'FP', 'min'),getTarget(scores, 'Trees', 'FP', 'min')]
bestFN= [getTarget(scores, 'Depth', 'FN', 'min'),getTarget(scores, 'Trees', 'FN', 'min')]
```

```python
#find the final best f1 score
bestScores = pd.DataFrame(columns=['Depth', 'Trees','Accuracy','F1','TP','TN','FP','FN'])
bestScores = bestScores.append(classifier(imputedTrain, trainLabels, imputedTest, testLabels, bestAcc[0], bestAcc[1])[0], ignore_index=True)
bestScores = bestScores.append(classifier(imputedTrain, trainLabels, imputedTest, testLabels, bestF1[0], bestF1[1])[0], ignore_index=True)
bestScores = bestScores.append(classifier(imputedTrain, trainLabels, imputedTest, testLabels, bestTP[0], bestTP[1])[0], ignore_index=True)
bestScores = bestScores.append(classifier(imputedTrain, trainLabels, imputedTest, testLabels, bestTN[0], bestTN[1])[0], ignore_index=True)
bestScores = bestScores.append(classifier(imputedTrain, trainLabels, imputedTest, testLabels, bestFP[0], bestFP[1])[0], ignore_index=True)
bestScores = bestScores.append(classifier(imputedTrain, trainLabels, imputedTest, testLabels, bestFN[0], bestFN[1])[0], ignore_index=True)
bestScores
```
