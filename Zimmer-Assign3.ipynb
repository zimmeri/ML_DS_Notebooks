```python
import math
import pandas as pd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sea
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score,  precision_recall_fscore_support, confusion_matrix, classification_report
```

```python
le = LabelEncoder()
modelData = pd.read_csv('master.csv')
modelData['sex'] = modelData['sex'].apply(lambda x: 1 if (x == 'male') else 0)
modelData['generation'] = le.fit(modelData['generation']).transform(modelData['generation'])
modelData['age'] = le.fit(modelData['age']).transform(modelData['age'])
modelData['country'] = le.fit(modelData['country']).transform(modelData['country'])
modelData['GDPyear'] = modelData[' gdp_for_year ($) '].apply(lambda x: float(x.replace(",", "")))
modelData['GDPcap'] = modelData.apply(lambda x: float(x['gdp_per_capita ($)']), axis=1)
modelData = modelData.drop(columns=['country-year',' gdp_for_year ($) ','gdp_per_capita ($)' , 'HDI for year'])
modelData.corr()
```

## 2. [10 pts] Explain in detail how one should set up the problem? Would it be a regression or a classification problem? Any unsupervised approach to see some patterns exist or not?
No matter what model type I use, I will be breaking it into a 70-30 training split. I am not planning on using crossvalidation because I'm not sure what hyperparameters I will need to adjust, but this may change later. <br>
I believe this will be a classification problem, but I am going to try both in order to verify this assumption. I think it is classification because we are trying to break the input down into groups in order to answer a binary question.<br>
In order to set up this problem for a yes/no output, I need to combine the data into groups and make a binary dependant variable. I group by country, year, and sex to find the groups GDP per year, GDP per capita, population, and suicides per 100 thousand. I make a binary dependant value called suicideCompared that determines if the groups suicide rate is above average. For the actual model, I will drop suicides per 100 thousand. <br>
For classification, I am going to use Random Forest since it will break the data down into high risk groups. For regression, I am going to use logistic regression.<br>
Below, I show a k-means cluster on the original data to see if there are any patterns. I will plot on a couple of different features. It seems like certain countries in certain years have above average suicide rates.

```python
countries = set(modelData['country'].values)
reduced = pd.DataFrame()
for country in countries:
    local = modelData[['country','year','GDPyear', 'GDPcap','sex']].groupby('country').get_group(country).groupby(['year','sex']).mean()
    local = local.merge(modelData[['country', 'year', 'sex', 'suicides/100k pop']].groupby('country').get_group(country).groupby(['year','sex']).sum().rename(columns={'suicides/100k pop':'suicides'}), on=['year','sex'])
    local = local.merge(modelData[['country', 'year', 'sex', 'population']].groupby('country').get_group(country).groupby(['year','sex']).sum(), on=['year','sex'])
    local['country'] = country
    reduced = reduced.append(local)
reduced = reduced.sort_values(['country','year'])
reduced['suicideCompared'] = reduced['suicides'].copy()
reduced['suicideCompared'] = np.where(reduced['suicideCompared'] < reduced['suicides'].mean(), 0, 1)
reduced = reduced.reset_index(drop=False).drop(columns=['country_y', 'country_x'])
reduced
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(modelData)
y_kmeans = kmeans.predict(modelData)
color= ['red' if l == 1 else 'green' for l in y_kmeans]
plt.scatter(modelData['country'], modelData['GDPyear'], s=50, color=color)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

## 6. [20 pts] Devise a classification problem and present a prototype model. Note that we will continue to this problem in the following modules.
I will be doing both classification and regression. As stated before, I will use a random forest and logistic regression.<br>
As you can see, the Logistic Regression model performed much worst in every category than the random forest. The random forest could definitely be improved as the scores are not that great.

```python
#encode country
reduced['country'] = le.fit(reduced['country']).transform(reduced['country'])
#split data
x = np.asarray(reduced.iloc[:,:6])
y = np.asarray(reduced.iloc[:,6])

#scale
scale = StandardScaler()
scale.fit(x)
scaleX = scale.transform(x)

#split
x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=.3)
```

### Logistic Regression

```python
lr = LogisticRegression(C=0.001, solver='liblinear').fit(x_train, y_train)

predictions = lr.predict(x_test)

print('Precision, recall, fscore, support\n', precision_recall_fscore_support(y_test, predictions, average='weighted'))

cm = confusion_matrix(y_test, predictions)
print('\nCM\n', cm)

print('\n', classification_report(y_test, predictions))
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=30, max_depth=20, min_samples_split=2, min_samples_leaf=5,
                                       max_leaf_nodes=100, max_features=6) 

rf.fit(x_train, y_train)

predictions = rf.predict(x_test)
ranks = pd.DataFrame([['year','GDPyear','GDPcap suicides','population','country'], rf.feature_importances_]).T

print('Training Accuracy', rf.score(x_train, y_train))
print('Testing Accuracy', rf.score(x_test, y_test))
print('\nCM\n', confusion_matrix(y_test, predictions))
print('\n', classification_report(y_test, predictions))
print(ranks.sort_values(by=1, ascending=False))
```
